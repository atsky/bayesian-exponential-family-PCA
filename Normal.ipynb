```python
import pymc3 as pm
import numpy as np
from numpy import random
import pylab as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
x = random.normal(loc=1.0, scale=2.0, size=200)
```

```python
with pm.Model() as model:
    # TODO
    m = pm.Normal('m',0, 1)
    sd = pm.Gamma('sd', 1, 1)
    
    x= pm.Normal('x', m, sd, observed=x)
    
    start = pm.find_MAP()
    step = pm.NUTS()
    trace = pm.sample(1000, start=start, step=step)
```

```python
pm.traceplot(trace)
```
